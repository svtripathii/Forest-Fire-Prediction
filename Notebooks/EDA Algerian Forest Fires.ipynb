## Algerian Forest Fires Dataset 
The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attributes and 1 output attribute (class)
The 244 instances have been classified into fire (138 classes) and not fire (106 classes) classes.

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
dataset= pd.read_csv('Algerian_forest_fires_dataset.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

## Data Cleaning

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
df.head()
```

```python
# df[['Region']]=df[['Region']].astype(int) Prob.: Unable to convert NaN Values so in next code we are dropping NaN Values
df[['Region']]=df[['Region']].dropna().astype(int)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip()
```

```python
df.info()
```

## Change the required columns as integer data type

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]= df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df[['Rain', 'FFMC',
       'DMC', 'DC', 'ISI', 'BUI', 'FWI']]= df[['Rain', 'FFMC',
       'DMC', 'DC', 'ISI', 'BUI', 'FWI']].astype(float)
```
